4. Removing high‐frequency components and reconstructing the image.
5. Removing low‐frequency components and reconstructing the image.
6. Comparing two images via a similarity index (to see if we can recognize the same person based on Fourier data).

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# For nicer plots inline with Jupyter:
%matplotlib inline
````

## Step 2: Load Your Selfie Image

Replace `'selfie.jpg'` with the path to your own image file.  

*Tip:* Make sure your image is in the same folder as this notebook (or provide the full path). If the image is large, consider resizing it to speed up processing.

````python
# Load an image
image_path = 'selfie.jpg' 
img = Image.open(image_path).convert('L')  # Convert to grayscale for simpler FFT

# Convert to numpy array
img_array = np.array(img)
print("Image shape:", img_array.shape)
print("Image maximum:", np.max(img_array))
print("Image minimum:", np.min(img_array))

# Display the image
plt.figure()
plt.title("Original Image")
plt.imshow(img_array)
plt.axis('off')
plt.show()
````

## Step 3: Perform the 2D FFT and Visualize the Frequency Domain

Here, we apply the 2D Fourier transform to our image. Mathematically, if f(x,y) is the value of our image at pixel coordinates (x,y) (with x ranging from 0 to M-1 and y from 0 to N-1), its 2D Fourier transform F(u,v) is given by:

$$
F(u, v) \;=\;
\sum_{x=0}^{M-1} \sum_{y=0}^{N-1}
f(x, y)\,
\exp\!\biggl(-\,2\pi i\!\Bigl(\frac{u\,x}{M} \;+\; \frac{v\,y}{N}\Bigr)\biggr).
$$

where (u,v) represents the spatial frequencies in the horizontal (x) and vertical (y) directions. In practice, we use np.fft.fft2 to compute this transform and then shift everything with reference to the lowest frequency component. We then take the log magnitude of our fourier transform to visualize it.


````python
# Compute 2D FFT
f = np.fft.fft2(img_array)
fshift = np.fft.fftshift(f)

# Compute the magnitude spectrum
magnitude_spectrum = np.log1p(np.abs(fshift))

# Show frequency domain
plt.figure()
plt.title("Magnitude Spectrum")
plt.imshow(magnitude_spectrum)
plt.axis('off')
plt.show()
````

## Step 4: Remove High‐Frequency Components

High-frequency components in images represent sharp features such as edges and fine details, we can smooth our image by removing these components:
1. Create a low‐pass mask that allows only the lower frequencies near the center.
2. Multiply the shifted FFT by this mask.
3. Invert the shift and do an inverse Fourier transform.

### Exercise
Change the radius, see how much higher frequency components can be removed.

````python
# 1. Create a low-pass filter mask
rows, cols = img_array.shape     # Get the dimensions of the image (number of rows, number of columns)
crow, ccol = rows // 2, cols // 2  # Compute the center (row index, column index)

radius = 10  # keep this many pixels around the center
mask = np.zeros((rows, cols), dtype=np.uint8)  # Start with a 2D array of zeros (same shape as the image)
y, x = np.ogrid[:rows, :cols]  # Create open grids for row and column indices
dist_from_center = np.sqrt((x - ccol)**2 + (y - crow)**2)  # Distance of each (x, y) from the center
mask[dist_from_center <= radius] = 1  # Set those points within 'radius' of center to 1 => “pass” those frequencies

# 2. Apply the mask
low_pass_fshift = fshift * mask  # Multiply the shifted FFT by the mask, zeroing out high-frequency components

# 3. Inverse shift and inverse FFT
low_pass_ishift = np.fft.ifftshift(low_pass_fshift)  # Undo the center shift
img_low_pass = np.fft.ifft2(low_pass_ishift)         # Compute the inverse FFT to return to the spatial domain
img_low_pass = np.abs(img_low_pass)                  # Take the magnitude (absolute value) of the complex output

# 4. Plot the result
plt.figure()
plt.title("Low-Pass Filtered Image (Removed High Frequencies)")
plt.imshow(img_low_pass)
plt.axis('off')
plt.show()
````

## Step 5: Remove Low‐Frequency Components

Similarly, we can remove the low‐frequency components by creating a high‐pass filter:
1. We’ll define a region around the center and set it to zero.
2. Invert shift and compute the inverse FFT.
What do you notice?

````python
# Create a high-pass filter mask (inverse of low-pass)
high_pass_mask = np.ones((rows, cols), dtype=np.uint8)
high_pass_mask[dist_from_center <= radius] = 0

# Apply the mask
high_pass_fshift = fshift * high_pass_mask

# Inverse shift and inverse FFT
high_pass_ishift = np.fft.ifftshift(high_pass_fshift)
img_high_pass = np.fft.ifft2(high_pass_ishift)
img_high_pass = np.abs(img_high_pass)

plt.figure()
plt.title("High-Pass Filtered Image (Removed Low Frequencies)")
plt.imshow(img_high_pass)
plt.axis('off')
plt.show()
````

````python
## Step 4: Denoising example!

# 1. Compute the FFT and shift
f = np.fft.fft2(img_array)
fshift = np.fft.fftshift(f)

# 2. Calculate the magnitude spectrum
magnitude_spectrum = np.abs(fshift)

# 3. Define a threshold; frequencies below this magnitude will be zeroed out
threshold = 1000000 # <-- adjust based on your image/noise
fshift_denoised = np.where(magnitude_spectrum < threshold, 0, fshift)

# 4. Inverse shift and inverse FFT
f_ishift = np.fft.ifftshift(fshift_denoised)
img_denoised = np.fft.ifft2(f_ishift)
img_denoised = np.abs(img_denoised)

# 5. Plot the denoised image
plt.figure()
plt.title("Denoised Image via FFT Thresholding")
plt.imshow(img_denoised)
plt.axis('off')
plt.show()
````

## Step 6: Similarity Index Between Fourier Transforms

Suppose we want to see if two images (say two selfies) belong to the same person by comparing their Fourier transforms.

One simple way: compute the correlation coefficient between their magnitude spectra.  

### Exercise
Investigate machine-learning methadologies that use FFT to help quantify images!

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the first image (already done, presumably)
img = Image.open('selfie.jpg').convert('L')
img_array = np.array(img)

# Compute FFT/magnitude for the first image
f1 = np.fft.fft2(img_array)
f1_shift = np.fft.fftshift(f1)
mag1 = np.log1p(np.abs(f1_shift))  # log(1 + magnitude)
mag1 = mag1.flatten()

# Load the second image and resize to match the first
img2 = Image.open('other_selfie.jpg').convert('L')
img2 = img2.resize(img.size)  # Ensure same dimensions as the first image
img2_array = np.array(img2)

# Compute FFT/magnitude for the second image
f2 = np.fft.fft2(img2_array)
f2_shift = np.fft.fftshift(f2)
mag2 = np.log1p(np.abs(f2_shift))
mag2 = mag2.flatten()

# Compute correlation coefficient (Pearson)
corr_coeff = np.corrcoef(mag1, mag2)[0, 1]
print("Similarity index (corr coeff):", corr_coeff)
````
